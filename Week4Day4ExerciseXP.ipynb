Introduction to Data Analysis

What is Data Analysis?

Data analysis is the process of collecting, cleaning, organizing, and examining data to gather useful information, identify patterns, and support decision-making. It involves using statistical methods and software tools to transform data into meaningful insights and conclusions.

Why is Data Analysis Important?

Data analysis plays an important role in society because organizations and individuals generate enormous amounts of data every day. Businesses use data to understand customers, improve products, reduce costs, and increase profits. Governments analyze data to improve public services, while healthcare organizations use it to improve patient care. Without data analysis, important trends and opportunities could be overlooked.

Three Areas Where Data Analysis Is Used

Healthcare

Hospitals analyze patient records to improve treatments, predict disease outbreaks, and optimize staffing.

Business and Marketing

Companies analyze customer purchasing behavior to personalize advertising, predict sales, and improve customer satisfaction.

Finance

Banks use data analysis to detect fraud, evaluate loan applications, manage investment portfolios, and assess financial risks.

| Data Source                                           | Type             | Explanation                                                                                                                                |
| ----------------------------------------------------- | ---------------- | ------------------------------------------------------------------------------------------------------------------------------------------ |
| A company's financial reports stored in an Excel file | **Structured**   | Excel stores data in rows and columns with a defined format, making it easy to organize and analyze.                                       |
| Photographs uploaded to a social media platform       | **Unstructured** | Images do not have a predefined table or sructure and require specialized tools for analysis.                                                |
| A collection of news articles on a website            | **Unstructured** | News articles consist primarily of free-form text rather than organized rows and columns.                                                  |
| Inventory data in a relational database               | **Structured**   | Relational databases store data in tables with defined columns, data types, and relationships.                                             |
| Recorded interviews from a market research study      | **Unstructured** | Audio recordings are not organized in a tabular format and typically require transcription or speech analysis before they can be analyzed. |
